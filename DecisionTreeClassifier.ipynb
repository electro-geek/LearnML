```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
from sklearn.datasets import load_iris
```

```python
iris = load_iris()

```

```python
#import as an datframe
import seaborn as sns
```

```python
df = sns.load_dataset('iris')
```

```python
#independetn features and dependent fetures
X = df.iloc[:,:-1]
y = iris.target
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.33, random_state=42)
```

```python
X_train
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
treemodel = DecisionTreeClassifier()
#by default it will become post pruning 
```

```python
treemodel.fit(X_train,y_train)
```

```python
from sklearn import tree
plt.figure(figsize=(15,10))
tree.plot_tree(treemodel, filled='True')
```

```python
#now with hyperparmater tuning
#by now setting max_depth
treemodel = DecisionTreeClassifier(max_depth=2)
#this is post pruning
```

```python
treemodel.fit(X_train,y_train)
```

```python
from sklearn import tree
plt.figure(figsize=(15,10))
tree.plot_tree(treemodel, filled='True')
```

```python
treemodel.predict(X_test)
```

```python
#prediction 
y_pred = treemodel.predict(X_test)
from sklearn.metrics import accuracy_score,classification_report
score = accuracy_score(y_pred,y_test)
print(score)
```

```python
 print(classification_report(y_pred,y_test))
```

# Decision Tree with Pre-Prunning

```python
#Decision Tree with Pre-Prunning
from sklearn.tree import DecisionTreeClassifier
```

```python
tree_model = DecisionTreeClassifier()
```

```python
#pre-pruning
#paramters
parameter={
 'criterion':['gini','entropy','log_loss'],
  'splitter':['best','random'],
  'max_depth':[1,2,3,4,5],
  'max_features':['auto', 'sqrt', 'log2']
    
}
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
cv = GridSearchCV(tree_model, param_grid = parameter, cv = 5, scoring = "accuracy")
```

```python
cv.fit(X_train, y_train)
```

```python
cv.best_params_
```

```python
cv.predict(X_test)
```

```python
y_pred = cv.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score, classification_report
```

```python
score = accuracy_score(y_pred, y_test)
```

```python
score
```

```python
print(classification_report(y_pred,y_test))
```
