```python
#housing price dataset
from sklearn.datasets import load_boston
```

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = load_boston()
```

```python
df
```

```python
#as you can see above the data set is in key value pair
#so we will  use pandas to convert this into a dataframe
pd.DataFrame(df.data)
```

```python
#now add the target feature name
dataset = pd.DataFrame(df.data)
dataset.columns = df.feature_names
dataset.head()
```

```python
#now I create my dependent feature
dataset['Price'] = df.target
#new column of name Price is added into the dataset which is similar to target
```

```python
dataset.head()
```

```python
#dif=viding the dataset into independent and dependent features
X = dataset.iloc[:,:-1]#
y = dataset.iloc[:,-1]
```

```python
#linear regression
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import cross_val_score
lin_reg = LinearRegression()
mse = cross_val_score(lin_reg,X,y,scoring='neg_mean_squared_error', cv=5)
print(mse)
mean_mse = np.mean(mse)
print(mean_mse)
```

```python
#Ridge Regression
from sklearn.linear_model import Ridge
from sklearn.model_selection import GridSearchCV
ridge = Ridge()

#Setting up hyperparmaters
params = {'alpha':[1e-15,1e-10,1e-8,1e-3,1e-2,1,5,10,20]}#it will check for every value and give the one with most accuracy
ridge_regressor = GridSearchCV(ridge, params, scoring = 'neg_mean_squared_error', cv = 10)
ridge_regressor.fit(X,y)
```

```python
print(ridge_regressor.best_params_)
print(ridge_regressor.best_score_)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

```python
#now doing with linear regression
from sklearn.linear_model import LinearRegression
lin_reg = LinearRegression()
mse = cross_val_score(lin_reg, X_train, y_train, scoring='neg_mean_squared_error', cv=5)
print(mse)
```

```python
X_train.head()
```

```python
lin_reg.fit(X_train,y_train)
predict = lin_reg.predict(X_test)
from sklearn.metrics import r2_score
r2_score(y_test, predict)
```

```python
from sklearn.linear_model import Ridge
rd = Ridge()
rd.fit(X_train,y_train)
rd_pred = rd.predict(X_test)
r2_score(y_test,rd_pred)
```

```python
from sklearn.linear_model import Lasso
from sklearn.model_selection import GridSearchCV
ls_reg = Lasso()
```

```python
lasso_regressor = GridSearchCV(ls_reg, params, scoring = 'neg_mean_squared_error', cv = 10)
lasso_regressor.fit(X,y)
```

```python
print(lasso_regressor.best_params_)
```

```python
print(lasso_regressor.best_score_)
```

```python
#doing lasso regression with train test split
lasso_regressor.fit(X_train,y_train)
lasso_predict = lasso_regressor.predict(X_test)
r2_score(lasso_predict,y_test)
```

```python
#doing ridge regression with train test split
from sklearn.linear_model import Ridge
from sklearn.model_selection import GridSearchCV
params = {'alpha':[1e-15,1e-10,1e-8,1e-3,1e-2,1,5,10,20]}
rdg = Ridge()
rg_regressor = GridSearchCV(rdg,params,scoring='neg_mean_squared_error',cv=10)
rg_regressor.fit(X_train,y_train)
rg_predict = rg_regressor.predict(X_test)
r2_score(rg_predict,y_test)
```

```python
from sklearn.datasets import load_breast_cancer
```

```python
df = load_breast_cancer()
```

```python
df
```

```python
#Independent Features
X = pd.DataFrame(df['data'],columns=df['feature_names'])
```

```python
X.head()
```

```python
#Dependent feature
y = pd.DataFrame(df['target'], columns=['target'])
y
```

```python
#checking whether dataset is balanced or imbalanced 
y['target'].value_counts()
```

```python
##Train test split
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

```python
params = [{'C':[1,5,10]},{'max_iter':[100,150]}]
```

```python
model1 = LogisticRegression(C=100,max_iter=100)
```

```python
from sklearn.model_selection import GridSearchCV
model = GridSearchCV(model1,param_grid=params,scoring='f1',cv=5)
```

```python
model.fit(X_train,y_train)
```

```python
y_pred = model.predict(X_test)
```

```python

model.best_score_
```

```python
model.best_params_
```

```python
y_pred
```

```python
from sklearn.metrics import confusion_matrix,classification_report
```

```python
confusion_matrix(y_test,y_pred)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred)
```

```python
print(classification_report(y_test,y_pred))
```
